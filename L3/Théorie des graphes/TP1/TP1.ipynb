### Exercice I.1 : Des graphes dans le plan

Dans ce premier exercice, il s'agit de produire des graphes définis *géométriquement*, et de les afficher au format PostScript (.ps).
Tout d'abord, définissons ce que signifie un graphe produit *géométriquement*. Dans notre cas, il s'agit d'un graphe $G=(V,E)$ dont les sommets sont des points d'un plan fini de dimension 300 par 300. Ainsi, $\forall v \in V, v=(x,y)$ avec $x,y \in [0, \cdots, 300]$.
À partir de cette définition, nous allons produire deux fonctions qui vont nous permettre de générer aléatoirement de tels graphes, puis de les visualiser :
- `get_alea_sommets(nb_sommets)` qui prend en paramètre un entier définissant le nombre de sommets à créer dans le graphe et retourne une liste de sommets (donc de couples d'entiers)
- `get_alea_aretes(sommets, distance)` qui prend en paramètre une liste de sommets obtenus via `get_alea_sommets` ainsi qu'un entier *distance*. L'idée est de connecter dans le graphe par une arête toutes les paires de sommets telles que la distance des sommets de ces paires sont inférieures au paramètre *distance*. Cette fonction retourne une liste de couple d'arêtes du type `[((u,v), (w,x)),...]` tel que `u`, `v`, `w` et `x` sont des sommets.

#### Pour vous aider

- la fonction `randint(mini, maxi)` du package random retourne un entier compris entre les paramètres mini et maxi
- la fonction `euclidean(u,v)` du package scipy.spatial.distance retourne la distance euclidienne entre deux couples de points
- la fonction d'affichage au format PostScript vous est fournie : `plot_ps(nom_fichier, taille_sommets, sommets, aretes)`. Elle prend en paramètre un nom de fichier dans lequel sera produite la visualisation, la taille des sommets (un entier) tels qu'ils seront affichés dans la visualisation, la liste de sommets produite avec `get_alea_sommets(nb_sommets)` et la liste d'arêtes produite avec `get_alea_aretes(sommets, distance)`


```python
from random import randint
from scipy.spatial.distance import euclidean

min=0
max=300

def get_alea_sommets(nb_sommets):
    ma_liste=[]
    for n in range(nb_sommets):
        ma_liste.append((randint(min,max),randint(min,max)))
    return ma_liste
    
    
sommets=get_alea_sommets(100)
print("Voici les sommets :\n")
print (sommets)
print("\nVoici les aretes :\n")

def get_distance(u,v):
    return euclidean(u,v)

def get_alea_aretes(sommets, distance):
    aretes=[]
    
    for i,sommetu in enumerate(sommets):
      for sommetv in sommets[i:]:
                   if get_distance(sommetu,sommetv) < distance:
                       aretes.append((sommetu,sommetv))
    return aretes

aretes=get_alea_aretes(sommets, 50)
print (aretes)
    
```

```python
def plot_ps(nom_fichier, taille_sommets, sommets, aretes):
    fichier=open(nom_fichier+".ps", "w")
    fichier.write("%!PS-Adobe-3.0\n%%BoundingBox: 0 0 300 300\n")
    for xu, yu in sommets:
        fichier.write(str(xu)+ " "+str(yu)+" "+str(taille_sommets)+" 0 360 arc\n0 setgray\nfill\nstroke\n")
    fichier.write("\n")
    for src, cible in aretes:
        fichier.write(str(src[0])+" "+str(src[1])+ " moveto\n")
        fichier.write(str(cible[0])+" "+str(cible[1])+ " lineto\nstroke\n")
    fichier.write("showpage\n")
    fichier.close()

plot_ps("Exemple", 3, sommets, aretes)
```

Cette liste d'adjacence prend la forme d'un dictionnaire avec pour clés les sommets, et pour valeurs la liste de voisins du sommet correspondant
- créer une fonction `degre_sommet_listeadjacence(liste_adjacence, sommet)` qui prend en paramètres la liste d'adjacebce ainsi qu'un sommet (un couple abscisse, ordonnée), et calcule le degré de ce sommet
- utiliser la fonction `getsizeof(objet)` du package *sys* pour mesurer la taille des structures de données

```python
def degre_sommet_listearetes(liste_aretes, sommet):
    degree =0
    for u,v in liste_aretes: #u=(x1,y1) v=(x2,y2)
        if u==sommet:
            degree+=1
        elif v== sommet:
            degree+=1
    return degree
            
degre_sommet_listearetes(aretes, sommets[6])
```

```python
def to_listeadjacence(liste_aretes, liste_sommets):
    adjacent=dict() #ajdacent={}
    for u in liste_sommets:
        adjacent[u]=[]
    for u,v in liste_aretes:
        adjacent[u].append(v)
        adjacent[v].append(u)
    return adjacent
                

adjacence=to_listeadjacence(aretes, sommets)
print(adjacence)
```

```python
class Sommet:
    def __init__(self,x,y):
        self._x=x
        self._y=y
        self._voisins = []
    
    @property
    def x(self):
        return self._x
    
    @x.setter
    def x(self, x): 
        self._x = x
    
    
    @property
    def y(self):
        return self._y
    
    
    @y.setter
    def y(self, y): 
        self._y = y
    
    
    
    def add_voisins(self,v):
        self._voisins.append(v)
     
    
    @property
    def voisins(self):
        return self._voisins
    
    @property
    def degre(self):
        return len(self._voisins)

```

```python
class Graphe:
    def __init__(self,V,E):
            self._V= V
            self._E= E

            
            
    @property 
    def V(self):
        return self._V
    
    @V.setter
    def V(self,sommets):
        self._V=sommets
        
    @property
    def E(self):
        return self._E
    
    @E.setter
    def E(self,aretes):
        self._E=aretes
    
    
    
    @property
    def n(self):
        return len(self._V)
    
    
    @property
    def m(self):
        return len(self._E)
    
    
    @property
    def max_nb_aretes(self):
        return (self.n *(self.n-1)/2)
```
