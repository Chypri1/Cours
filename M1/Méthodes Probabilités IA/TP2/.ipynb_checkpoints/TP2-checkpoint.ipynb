```python
import pandas as pd
def read_bio(fn):
    data = []
    sentence_idx = 0
    with open(fn,'r',encoding = 'UTF-8') as f:
        for line in f.readlines():
            if line == '\n':
                sentence_idx += 1
            else:
                data.append([sentence_idx]+line.strip().split(" "))
    return pd.DataFrame(data,columns = ['sentence','word','label'])
```

```python
nomFichier = "train.ester1.cut.bio"
train_df = read_bio(nomFichier)
train_df
```

```python
A = pd.DataFrame(0, index=train_df['label'].unique(), columns=train_df['label'].unique())
```

```python
for idx, sentence in train_df.groupby('sentence'):
    labels = sentence['label'].to_list()
    for i in range(len(labels)-1):
        A.at[labels[i], labels[i+1]]+=1

A
```

```python
A = A.div(A.sum(axis=1),axis=0)
# A = matrice de transition
A
```

```python
def matrice_mot(fichier):
    # Calculer le nombre total de lignes
    total_rows = len(fichier)

    # Grouper par "label" et "word", puis compter le nombre d'occurrences de chaque mot pour chaque label
    word_counts = fichier.groupby(['label', 'word']).size().reset_index(name='count')

    # Calculer la probabilité d'apparition de chaque mot pour chaque label
    word_counts['probability'] = word_counts['count'] / total_rows

    # Convertir le DataFrame en un dictionnaire de la forme dict[label] = dict[word] = probabilité
    result_dict = {}
    for label, word, probability in zip(word_counts['label'], word_counts['word'], word_counts['probability']):
        if label not in result_dict:
            result_dict[label] = {}
        result_dict[label][word] = probability
    return result_dict

B=matrice_mot(train_df)
# B = matrice d'émission 
B
```
